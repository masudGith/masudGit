```python
import openpyxl
import datetime

# Get the current day's name
current_day = datetime.datetime.now().strftime('%A')

# Specify the path to your Excel file
excel_file_path = r'D:\Python\MAssudurinment\Excel.xlsx'

# Load the Excel workbook
workbook = openpyxl.load_workbook(excel_file_path)

# Get the worksheet corresponding to the current day's name
worksheet = workbook[current_day]

# # Create a dictionary to store keyword ranges
# keyword_ranges = {}

# # Maximum number of cells to check in each column
# max_cells_to_check = 50

# # Iterate through columns and check the first max_cells_to_check cells
# for col_index in range(1, worksheet.max_column + 1):
#     keyword_column = None
    
#     # Iterate through the cells in the column up to max_cells_to_check
#     for row_index in range(1, max_cells_to_check + 1):
#         cell = worksheet.cell(row=row_index, column=col_index)
#         cell_value = cell.value
#         if cell_value and "keyword" in str(cell_value).lower():
#             keyword_column = col_index
#             break

# #     if keyword_column is not None:
# #         # Store the range of cells containing the keyword name and value
# #         keyword_ranges[keyword_column] = (2, row_index + 1, keyword_column, keyword_column + 1)

# #         # Extract keywords and values from the identified column
# #         for row_index in range(2, worksheet.max_row + 1):
# #             keyword_name = worksheet.cell(row=row_index, column=keyword_column).value
# #             keyword_value = worksheet.cell(row=row_index, column=keyword_column + 1).value
            
# #             if keyword_name and keyword_value:
# #                 keywords[keyword_name] = keyword_value

# #         break  # Stop iterating after finding the "Keywords" column

# Create a list to store keyword ranges
keyword_ranges = []

# Maximum number of cells to check in each column
max_cells_to_check = 50

# Iterate through columns and check the first max_cells_to_check cells
for col_index in range(1, worksheet.max_column + 1):
    keyword_column = None
    
    # Iterate through the cells in the column up to max_cells_to_check
    for row_index in range(1, max_cells_to_check + 1):
        cell = worksheet.cell(row=row_index, column=col_index)
        cell_value = cell.value
        if cell_value and "keyword" in str(cell_value).lower():
            keyword_column = col_index
            break

    if keyword_column is not None:
        # Find the starting and ending row indices for the keyword values
        start_row = None
        end_row = None
        for row_index in range(2, worksheet.max_row + 1):
            if worksheet.cell(row=row_index, column=keyword_column + 1).value:
                if start_row is None:
                    start_row = row_index
                end_row = row_index

        # Construct the cell range string
        keyword_ranges.append(f"{worksheet.cell(row=start_row, column=keyword_column + 1).coordinate}:{worksheet.cell(row=end_row, column=keyword_column + 1).coordinate}")

        break  # Stop iterating after finding the "Keywords" column

if keyword_ranges:
    # Print the extracted keyword value ranges serially
    print("Keyword Value Ranges:")
    for keyword_range in keyword_ranges:
        print(keyword_range)
else:
    print("Keyword column not found in the worksheet.")

# Close the workbook
workbook.close()
```
